```python
import seaborn as sns
import numpy as np

# Generating example data for three groups
# 'TOT2UR', 'TOT3UF', 'TOT4NBDM', 'TOT5NBDR', 'TOT6NBDF',
data1 = data['TOT1UM']*60/1000
data2 = data['TOT2UR']*60/1000
data3 = data['TOT3UF']*60/1000
data4 = data['TOT4NBDM']*60/1000
data5 = data['TOT5NBDR']*60/1000
data6 = data['TOT6NBDF']*60/1000

# Combining the data into a single DataFrame for Seaborn
x = np.concatenate([data1, data2, data3, data4, data5, data6])
group = ['Group 1'] * 5338 + ['Group 2'] * 5338 + ['Group 3'] * 5338 + ['Group 4'] * 5338 + ['Group 5'] * 5338 + ['Group 6'] * 5338
df = pd.DataFrame({'Data': x, 'Group': group})

# Creating multiple box plots using Seaborn
plt.figure(figsize=(14, 6))
sns.boxplot(x='Group', y='Data', data=df)
plt.title('Multiple Box Plots using Seaborn')
plt.show()
```

```python
selected=['TOT6NBDF', 'SR_MAX', 'SR_MIN', 'WS_MAX', 'WS_AVE',
       'WD_AVE', 'WD_SD']

temp=data[selected]
cor=temp.corr()
plt.figure(figsize=(8, 6))  # Set the size of the heatmap
sns.heatmap(cor, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Pearson Correlation Coefficients Heatmap')
plt.show()
```
